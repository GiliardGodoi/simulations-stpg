```python

# search_key_path(graph, terminals, start)
def key_paths(graph, terminals, start):
    visited = set()
    visited.add(start)
    def search(s, v, subgraph):
        path = [s, v]
        stack = list()
        # print('--->', s, v)
        if v not in terminals \
            and subgraph.degree(v) <= 2:
            stack.append(v)

        while stack:
            u = stack.pop()
            visited.add(u)
            for w in subgraph.adjacent_to(u):
                if (w not in visited) \
                    and (w not in terminals) \
                    and (subgraph.degree(w) <= 2):
                    stack.append(w)
                    path.append(w)
                elif w not in visited and v != u:
                    # print(stack, u, w)
                    path.append(w)
                # elif len(stack) == 0:
                #     path.append(w)

        return path


    paths = list()
    
    for v in graph.adjacent_to(start):
        if v not in visited:
            path = search(start, v, graph)
            paths.append(path)
    return paths
```

```python
paths = key_paths(graph, terminals, 5)

print(paths)

assert [5, 2] in paths
assert [5, 7, 9, 1] in paths
```

```python
def key_path2(graph, terminals, start, finish):
    def P(v, u, path=None):
        if path is None:
            path = [v]
        
        path.append(u)

        if u in terminals \
            or graph.degree(u) > 2:
            return path
        else:
            for w in graph.adjacent_to(u):
                if w != v:
                    return P(u, w, path=path)
    return P(start, finish)
```
